```python
import pandas as pd
```

```python
players = pd.read_csv("players.csv")
players['birth_year'] = pd.to_numeric('19' + players['birth_date'].str.strip().str[-2:])
players.drop(['column_a', 'url', 'birth_date'], axis=1, inplace=True)
players = players[pd.notnull(players['height'])]
```

```python
height_ft = players['height'].str.strip().str[0:1].astype(int)
height_in = players['height'].str.strip().str[2:4].astype(int)

players['height'] = pd.to_numeric(height_ft * 12 + height_in)
```

```python
players['ncaa_efgpct'] = round((players['ncaa_fgpg'] + 0.5 * players['ncaa_3ptpg']) / players['ncaa_fgapg'], 3)
players['career_length'] = players['active_to'] - players['active_from']
players['bmi'] = round((players['weight']/(players['height'])**2)*703, 1)
players['age_entering_league'] = players['active_from'] - players['birth_year']
players = players[players['age_entering_league'].notna()]
players['age_entering_league'] = players['age_entering_league'].astype(int)
```

```python
players['position'] = players['position'].str.strip().str[0]
```
